```python
a=[20,10,90]
b=[40,4,100]
c=[20,25,5]
    
t = 0

def min_min(aa , bb):
    if aa<bb:
        return aa , 0
    
    return bb , 1

did=[1 for value in range(0,n)]

min_time_idx = 1

while sum(did) != 0:
    
    if min_time_idx == 0:
        ch = a
    else:
        ch = b
        
    for j in range(0,n):
        if ch[j] < ch[min_time_idx] and did[j] != 0:
            min_time_idx = j
    print(min_time_idx)
    did[min_time_idx] = 0
    
    if min_time_idx == 0:
        aaa , min_time_idx = min_min(a[min_time_idx] , b[min_time_idx]+c[min_time_idx])
        t = t + aaa
    else:
        bbb , min_time_idx = min_min(b[min_time_idx] , a[min_time_idx]+c[min_time_idx] )
        t = t + bbb

print(t)
```
